```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
balance_scale = fetch_ucirepo(id=12) 
  
# data (as pandas dataframes) 
X1 = balance_scale.data.features 
y1 = balance_scale.data.targets 
  
# metadata 
print(balance_scale.metadata) 
  
# variable information 
print(balance_scale.variables) 
```

Classificação:

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, KFold
import numpy as np
from sklearn.preprocessing import MinMaxScaler

normalizador1 = MinMaxScaler(feature_range=(0, 1))
x_norm1 = normalizador1.fit_transform(X1)

c1 = np.array([0.1, 1, 10, 100])
kernel1 = ['linear', 'poly', 'rbf', 'sigmoid']
polinomio1 = np.array([2, 3, 4])

valores_grid1 = {'C':c1,'kernel': kernel1,'degree':polinomio1}

modelo1 = SVC()
kfold1 = KFold(n_splits=3, shuffle=True, random_state=42) 
gridsvm1 = GridSearchCV(estimator=modelo1, param_grid=valores_grid1, cv=kfold1, n_jobs=-1, scoring='r2', verbose=2)
gridsvm1.fit(x_norm1, y1.values.ravel())

print('Melhor R2:', gridsvm1.best_score_)
print('Melhores Hiperparâmetros:', gridsvm1.best_params_)
```

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
concrete_compressive_strength = fetch_ucirepo(id=165) 
  
# data (as pandas dataframes) 
X = concrete_compressive_strength.data.features 
y = concrete_compressive_strength.data.targets 
  
# metadata 
print(concrete_compressive_strength.metadata) 
  
# variable information 
print(concrete_compressive_strength.variables) 
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, KFold
import numpy as np
from sklearn.preprocessing import MinMaxScaler

normalizador = MinMaxScaler(feature_range=(0,1))
x_norm = normalizador.fit_transform(X)

c = np.array([0.1, 1, 10, 100]) 
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
polinomio = np.array([2, 3, 4]) 
epsilon = np.array([0.01, 0.1, 1])

valores_grid = {'C':c,'kernel': kernel,'degree':polinomio, 'epsilon':epsilon}

modelo = SVR()
kfold = KFold(n_splits=3,shuffle=True)

gridsvm = GridSearchCV(estimator=modelo, param_grid=valores_grid, cv=kfold, n_jobs=-1, scoring='r2', verbose=2) 
gridsvm.fit(x_norm,y.values.ravel())

print('Melhor R2:', gridsvm.best_score_)
print('Melhores Hiperparâmetros:', gridsvm.best_params_)
```
